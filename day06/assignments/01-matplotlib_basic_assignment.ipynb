## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**

## Exercise 5

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
profit=table["total_profit"].tolist()
facecream=table["facecream"].tolist()
facewach=table["facecream"].tolist()
toothbath=table["toothpaste"].tolist()
bathingsoap=table["bathingsoap"].tolist()
shampoo=table["shampoo"].tolist()
moisturizer=table["moisturizer"].tolist()
total=[]

tfacecream=0
for i in range(len(profit)):
    tfacecream+=facecream[i]
total.append(tfacecream)
tfacewash=0
for i in range(len(profit)):
    tfacewash+=facewash[i]
total.append(tfacewash)
ttoothbath=0
for i in range(len(profit)):
    ttoothbath+=toothbath[i]
total.append(ttoothbath)

tbathingsoap=0
for i in range(len(profit)):
    tbathingsoap+=bathingsoap[i]
total.append(tbathingsoap)

tshampoo=0
for i in range(len(profit)):
    tshampoo+=shampoo[i]
total.append(tshampoo)

tmoisturizer=0
for i in range(len(profit)):
    tmoisturizer+=moisturizer[i]
total.append(tmoisturizer)

mylabels=["facecream" ,"facewash" ,"toothpaste" ,"bathingsoap" ,"shampoo" ,"moisturizer"]
print(total)
print(mylabels)

plt.pie(total, labels = mylabels,autopct='%1.2f%%')
plt.show()
```
